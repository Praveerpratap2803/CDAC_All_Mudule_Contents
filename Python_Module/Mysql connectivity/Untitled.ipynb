```python
cnx = mysql.connector.connect(user="root",
                              password="root",
                              host="127.0.0.1",
                             database='XYZ'
                             )
cur=cnx.cursor()
cur.execute("""CREATE TABLE student (roll_no int,name varchar(20),marks int );""")
cnx.commit()
cnx.close
```

```python
cnx = mysql.connector.connect(user="root",
                              password="root",
                              host="127.0.0.1",
                              database='XYZ')
cur = cnx.cursor()
cur.execute("""select * from student""")
for i in cur:
    print(i)
cnx.commit()
cnx.close()
```

```python
data = [
    [9,"name1",91],
    [8,"name2",92]
]
cnx = mysql.connector.connect(user="root",
                              password="root",
                              host="127.0.0.1",
                              database="XYZ"
)
```

```python
cur = cnx.cursor()
for i in data:
    a="INSERT INTO student(roll_no,name,marks) values("+str(i[0])+", '"+i[1]+"',"+str(i[2])+");"
    cur.execute(a)
cnx.commit()
cnx.close()
```
