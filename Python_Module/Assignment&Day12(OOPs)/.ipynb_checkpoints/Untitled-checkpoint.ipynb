```python
class car:
    def __init__ (self,color,price,avg,type_fuel):
        self.color=color
        self.price=price
        self.avg=avg
        self.type_fuel=type_fuel
    def display(self):
        print(self.avg,self.color,self.price)
c1 = car("white",12,13,"type1")
c1.display()
```

```python
class student:
    def __init__(self,m1,m2):
        self.m1=m1
        self.m2=m2
    def display(self):
        print(self.m1,self.m2)

s1 = student(10,20)
s2=student(100,200)
s1.display()
        
```
